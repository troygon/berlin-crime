# 04 Create subsets

```python
# Create two subsets for crime rate where the District Total is separate from the rest of the data

# Subset with rows containing '-Total'
district_crime = berlin_crime[berlin_crime['district'].str.contains('-Total', na=False)]

# Subset with rows without '-Total'
neighborhood_crime = berlin_crime[~berlin_crime['district'].str.contains('-Total', na=False)]

# Subset for whole berlin
berlin_crime_subset = berlin_crime[berlin_crime['district'].str.contains('Berlin', na=False)]
```

```python
berlin_crime_subset.head()
```

```python
district_crime.head()
```

```python
neighborhood_crime.head()
```

```python
# Remove '-Total' in the district column for the kiez subset
district_crime.loc[:, 'district'] = district_crime['district'].str.replace('-Total', '', regex=False)
```

```python
district_crime.tail(20)
```

```python
# Filter for Male and Female Population within the 2013-2022
gender_pop = berlin_pop_gender[
    (berlin_pop_gender['Gender_Nationality'].isin(['Male', 'Female'])) &
    (berlin_pop_gender['Year'].between(2013, 2022))
]
```

```python
gender_pop.head(20)
```

```python
# Filter for German and Foreigner Population within 2013-2022
nationality_pop = berlin_pop_gender[
    (berlin_pop_gender['Gender_Nationality'].isin(['German', 'Foreigner'])) &
    (berlin_pop_gender['Year'].between(2013, 2022))
]
```

```python
nationality_pop.head(20)
```

```python
migration_pop = berlin_pop_mig[
    (berlin_pop_mig['Migration_background'].isin(['Germans without migration background','Germans with migration background', 'Foreigners'])) &
    (berlin_pop_mig['Year'].between(2013, 2022))
]
```

```python
migration_pop.head(20)
```

```python
# Change dtypes
migration_pop = migration_pop.astype({'Population': 'int'})
```

```python
migration_pop.head()
```
